```python
# Standard EfficientNet for Fruits 360 Classification (No Attention Mechanism)
# Universidad Militar Nueva Granada - Deep Learning Course

import tensorflow as tf
from tensorflow.keras import layers, models, applications, callbacks, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import os
import time
import datetime

# Constants
IMG_HEIGHT, IMG_WIDTH = 100, 100
BATCH_SIZE = 32
EPOCHS = 30
LEARNING_RATE = 0.001
FREEZE_LAYERS = 100  # Number of layers to freeze in EfficientNetB0

print("TensorFlow version:", tf.__version__)

# Function to create standard EfficientNet model without attention mechanisms
def create_standard_efficientnet(input_shape=(IMG_HEIGHT, IMG_WIDTH, 3), num_classes=131):
    """
    Creates a standard EfficientNetB0 model with transfer learning but no attention mechanisms.
    """
    # Load pre-trained EfficientNetB0
    base_model = applications.EfficientNetB0(
        weights='imagenet',
        include_top=False,
        input_shape=input_shape
    )

    # Freeze early layers
    for layer in base_model.layers[:FREEZE_LAYERS]:
        layer.trainable = False

    # Create model with standard classification head
    inputs = base_model.input
    x = base_model.output

    # Add classification head (same structure as in attention model but without CBAM)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(1024, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs=inputs, outputs=outputs)

    # Compile model
    model.compile(
        optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Function to prepare data generators
def prepare_data_generators():
    # Data augmentation for training
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        validation_split=0.2  # 20% for validation
    )

    # Just rescaling for testing
    test_datagen = ImageDataGenerator(rescale=1./255)

    # Training generator
    train_generator = train_datagen.flow_from_directory(
        'Fruit-Images-Dataset-master/Training',
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='training'
    )

    # Validation generator
    validation_generator = train_datagen.flow_from_directory(
        'Fruit-Images-Dataset-master/Training',
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='validation'
    )

    # Test generator
    test_generator = test_datagen.flow_from_directory(
        'Fruit-Images-Dataset-master/Test',
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )

    return train_generator, validation_generator, test_generator

# Function to plot confusion matrix
def plot_confusion_matrix(cm, class_labels, num_classes_to_show=20):
    # Select the top N classes if there are too many
    if len(class_labels) > num_classes_to_show:
        class_representation = np.sum(cm, axis=0) + np.sum(cm, axis=1)
        top_classes_idx = np.argsort(class_representation)[-num_classes_to_show:]

        filtered_cm = cm[np.ix_(top_classes_idx, top_classes_idx)]
        filtered_labels = [class_labels[i] for i in top_classes_idx]
    else:
        filtered_cm = cm
        filtered_labels = list(class_labels.values())

    plt.figure(figsize=(12, 10))
    sns.heatmap(filtered_cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=filtered_labels, yticklabels=filtered_labels)
    plt.title('Standard EfficientNet - Confusion Matrix (Top Classes)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig('standard_efficientnet_confusion_matrix.png')
    plt.show()

# Main function
def main():
    # Make sure the dataset is available
    if not os.path.exists('Fruit-Images-Dataset-master'):
        print("Dataset not found. Please download and extract the Fruits 360 dataset first.")
        print("You can use the following commands:")
        print("!wget -O fruits-360.zip https://github.com/Horea94/Fruit-Images-Dataset/archive/refs/heads/master.zip")
        print("!unzip -q fruits-360.zip")
        return

    # Create directory for model checkpoints
    os.makedirs('standard_model_checkpoints', exist_ok=True)

    # Prepare data generators
    print("Preparing data generators...")
    train_generator, validation_generator, test_generator = prepare_data_generators()

    # Get number of classes
    num_classes = len(train_generator.class_indices)
    print(f"Number of fruit classes: {num_classes}")

    # Create standard EfficientNet model
    print("Creating standard EfficientNet model...")
    model = create_standard_efficientnet(
        input_shape=(IMG_HEIGHT, IMG_WIDTH, 3),
        num_classes=num_classes
    )

    # Print model summary
    model.summary()

    # Create timestamp for model identification
    timestamp = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

    # Setup callbacks
    callbacks_list = [
        # Early stopping
        callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=7,
            restore_best_weights=True
        ),
        # Learning rate reduction
        callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=1e-6
        ),
        # Model checkpoint
        callbacks.ModelCheckpoint(
            filepath=f'standard_model_checkpoints/standard_efficientnet_{timestamp}.h5',
            monitor='val_accuracy',
            save_best_only=True,
            mode='max'
        )
    ]

    # Train the model
    print("Training standard EfficientNet model...")
    start_time = time.time()

    # Calculate steps
    steps_per_epoch = train_generator.samples // train_generator.batch_size
    validation_steps = validation_generator.samples // validation_generator.batch_size

    history = model.fit(
        train_generator,
        steps_per_epoch=steps_per_epoch,
        epochs=EPOCHS,
        validation_data=validation_generator,
        validation_steps=validation_steps,
        callbacks=callbacks_list
    )

    standard_training_time = time.time() - start_time
    print(f"Standard EfficientNet training completed in {standard_training_time:.2f} seconds")

    # Evaluate on test set
    print("Evaluating standard EfficientNet model...")
    test_steps = (test_generator.samples // test_generator.batch_size) + 1
    standard_test_loss, standard_test_accuracy = model.evaluate(
        test_generator,
        steps=test_steps
    )
    print(f"Standard EfficientNet - Test accuracy: {standard_test_accuracy:.4f}")
    print(f"Standard EfficientNet - Test loss: {standard_test_loss:.4f}")

    # Plot training history
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Standard EfficientNet - Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Standard EfficientNet - Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.tight_layout()
    plt.savefig('standard_efficientnet_training_history.png')
    plt.show()

    # Generate confusion matrix
    print("Generating predictions and confusion matrix...")
    predictions = model.predict(
        test_generator,
        steps=test_steps
    )

    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = test_generator.classes[:len(predicted_classes)]
    class_labels = {v: k for k, v in test_generator.class_indices.items()}

    # Create classification report
    report = classification_report(
        true_classes,
        predicted_classes,
        target_names=list(class_labels.values()),
        output_dict=False
    )
    print("Standard EfficientNet - Classification Report:")
    print(report)

    # Save report to file
    with open(f'standard_efficientnet_report_{timestamp}.txt', 'w') as f:
        f.write(report)

    # Create and plot confusion matrix
    cm = confusion_matrix(true_classes, predicted_classes)
    plot_confusion_matrix(cm, class_labels)

    # Save the final model
    model.save(f"standard_efficientnet_model_{timestamp}.h5")
    print(f"Model saved as standard_efficientnet_model_{timestamp}.h5")

    # Print final message for comparison
    print("\nStandard EfficientNet Results Summary:")
    print(f"Test Accuracy: {standard_test_accuracy:.4f}")
    print(f"Training Time: {standard_training_time:.2f} seconds")
    print("\nYou can now compare these results with your Attention-Augmented EfficientNet model")
    print("Expected improvement from attention mechanisms: 1-3% in accuracy and better focus on relevant features")

# Run the main function
if __name__ == "__main__":
    main()
```
