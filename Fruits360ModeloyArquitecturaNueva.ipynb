```python
# Implementation Steps for Fruits 360 Classification with Attention-Augmented EfficientNet
# Universidad Militar Nueva Granada - Deep Learning Course

"""
This implementation guide provides step-by-step instructions and explanations
for implementing the Attention-Augmented EfficientNet model for fruit classification.
"""

# Step 1: Setup the environment and import dependencies
# These imports need to be executed at the beginning of your Google Colab notebook
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, applications, callbacks, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Conv2D, BatchNormalization, Activation, multiply, Reshape, Permute, Lambda
import zipfile
from tensorflow.keras import backend as K
import time
import datetime
from google.colab import drive
from sklearn.metrics import confusion_matrix, classification_report

# Optional: Mount Google Drive if you want to save your model there
# drive.mount('/content/drive')

# Step 2: Download and extract the Fruits 360 dataset
# Run this code to download and extract the dataset
!wget -O fruits-360.zip https://github.com/Horea94/Fruit-Images-Dataset/archive/refs/heads/master.zip
!unzip -q fruits-360.zip

# Define paths to training and test directories
train_dir = 'Fruit-Images-Dataset-master/Training'
test_dir = 'Fruit-Images-Dataset-master/Test'

# Step 3: Explore the dataset
# Print the number of classes and examples
print("Classes in training set:", len(os.listdir(train_dir)))
print("Classes in test set:", len(os.listdir(test_dir)))

# Count images in a few sample classes
sample_classes = list(os.listdir(train_dir))[:5]
for cls in sample_classes:
    train_count = len(os.listdir(os.path.join(train_dir, cls)))
    test_count = len(os.listdir(os.path.join(test_dir, cls)))
    print(f"Class {cls}: {train_count} training images, {test_count} test images")

# Step 4: Define the Attention Modules
# These are the key innovations in our model

# Squeeze-and-Excitation block as a custom layer
class SqueezeExciteBlock(layers.Layer):
    """
    Implements the Squeeze-and-Excitation block for channel attention as a custom Keras layer.

    This block helps the model focus on the most important feature channels.
    """
    def __init__(self, ratio=16, **kwargs):
        super(SqueezeExciteBlock, self).__init__(**kwargs)
        self.ratio = ratio
        self.gap = GlobalAveragePooling2D()

    def build(self, input_shape):
        filters = input_shape[-1]
        self.reshape = Reshape((1, 1, filters))
        self.dense1 = Dense(filters // self.ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
        self.dense2 = Dense(filters, activation='sigmoid', kernel_initializer='he_normal', use_bias=False)
        super(SqueezeExciteBlock, self).build(input_shape)

    def call(self, input_tensor):
        # Squeeze operation (global average pooling)
        se = self.gap(input_tensor)
        se = self.reshape(se)

        # Excitation operation (two FC layers with bottleneck)
        se = self.dense1(se)
        se = self.dense2(se)

        # Scale the input
        return multiply([input_tensor, se])

    def compute_output_shape(self, input_shape):
        return input_shape

def squeeze_excite_block(input_tensor, ratio=16):
    return SqueezeExciteBlock(ratio=ratio)(input_tensor)

# Spatial Attention module as a custom layer
class SpatialAttention(layers.Layer):
    """
    Implements the Spatial Attention module as a custom Keras layer.

    This module helps the model focus on important spatial locations in the feature maps.
    """
    def __init__(self, **kwargs):
        super(SpatialAttention, self).__init__(**kwargs)
        self.conv = Conv2D(1, kernel_size=7, padding='same', activation='sigmoid')

    def build(self, input_shape):
        super(SpatialAttention, self).build(input_shape)

    def call(self, input_tensor):
        # Apply average pooling and max pooling along channel axis
        avg_pool = tf.reduce_mean(input_tensor, axis=3, keepdims=True)
        max_pool = tf.reduce_max(input_tensor, axis=3, keepdims=True)

        # Concatenate pooled features
        concat = tf.concat([avg_pool, max_pool], axis=3)

        # Apply convolution to generate spatial attention map
        spatial_map = self.conv(concat)

        # Apply spatial attention
        return multiply([input_tensor, spatial_map])

    def compute_output_shape(self, input_shape):
        return input_shape

def spatial_attention(input_tensor):
    return SpatialAttention()(input_tensor)

# Convolutional Block Attention Module (CBAM) as a custom layer
class CBAM(layers.Layer):
    """
    Combines both channel and spatial attention mechanisms in a custom Keras layer.

    This is our key innovation for improving the model's focus on relevant features.
    """
    def __init__(self, ratio=16, **kwargs):
        super(CBAM, self).__init__(**kwargs)
        self.se_block = SqueezeExciteBlock(ratio=ratio)
        self.spatial_att = SpatialAttention()

    def call(self, input_tensor):
        # Apply channel attention
        channel_attention = self.se_block(input_tensor)

        # Apply spatial attention
        spatial_att = self.spatial_att(channel_attention)

        return spatial_att

    def compute_output_shape(self, input_shape):
        return input_shape

def cbam_block(input_tensor, ratio=16):
    return CBAM(ratio=ratio)(input_tensor)

# Step 5: Create the Attention-Augmented EfficientNet model
def create_attention_augmented_efficientnet(input_shape=(100, 100, 3), num_classes=131):
    """
    Creates our novel Attention-Augmented EfficientNetB0 model.

    This model enhances EfficientNet with CBAM attention mechanisms to improve
    feature focus and classification performance.

    Args:
        input_shape: Shape of input images
        num_classes: Number of classes to classify

    Returns:
        The compiled model
    """
    # Use EfficientNetB0 as the base model
    base_model = applications.EfficientNetB0(weights='imagenet', include_top=False, input_shape=input_shape)

    # Freeze early layers to prevent overfitting and speed up training
    for layer in base_model.layers[:100]:
        layer.trainable = False

    # Get the output from the base model
    x = base_model.output

    # Add CBAM attention module
    x = cbam_block(x)

    # Add global average pooling
    x = GlobalAveragePooling2D()(x)

    # Add dropout to prevent overfitting
    x = Dropout(0.5)(x)

    # Add dense layer with L2 regularization
    x = Dense(1024, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
    x = Dropout(0.3)(x)

    # Output layer
    predictions = Dense(num_classes, activation='softmax')(x)

    # Create the full model
    model = models.Model(inputs=base_model.input, outputs=predictions)

    return model

# Step 6: Data preparation with augmentation
# Set image dimensions and batch size
img_height, img_width = 100, 100
batch_size = 32

# Data augmentation for training images
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2  # 20% of training data for validation
)

# Only rescaling for testing images
test_datagen = ImageDataGenerator(rescale=1./255)

# Training generator
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

# Validation generator
validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Test generator
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Step 7: Create custom learning rate scheduler
def create_lr_scheduler():
    """
    Creates a learning rate scheduler with warm-up and decay phases.

    Returns:
        LearningRateScheduler callback
    """
    def lr_schedule(epoch, lr):
        if epoch < 3:  # Warm-up phase
            return lr * 1.2
        elif epoch < 15:
            return lr
        elif epoch < 25:
            return lr * 0.5
        else:
            return lr * 0.1
    return callbacks.LearningRateScheduler(lr_schedule)

# Step 8: Define callbacks for training
# Create EarlyStopping callback
early_stopping = callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

# Create ModelCheckpoint callback
timestamp = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
checkpoint_path = f"model_checkpoints/fruits_model_{timestamp}.h5"
os.makedirs(os.path.dirname(checkpoint_path), exist_ok=True)
checkpoint_callback = callbacks.ModelCheckpoint(
    checkpoint_path,
    monitor='val_accuracy',
    save_best_only=True,
    mode='max'
)

# Step 9: Create and compile the model
# Get number of classes
num_classes = len(train_generator.class_indices)
print(f"Number of fruit classes: {num_classes}")

# Create the model
model = create_attention_augmented_efficientnet(
    input_shape=(img_height, img_width, 3),
    num_classes=num_classes
)

# Compile the model
model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
model.summary()

# Step 10: Train the model
# Start training
epochs = 30
start_time = time.time()

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size,
    callbacks=[
        create_lr_scheduler(),
        early_stopping,
        checkpoint_callback
    ]
)

end_time = time.time()
training_time = end_time - start_time
print(f"Training completed in {training_time:.2f} seconds")

# Step 11: Evaluate the model
# Evaluate on test data
test_steps = test_generator.samples // test_generator.batch_size + 1
test_loss, test_accuracy = model.evaluate(test_generator, steps=test_steps)
print(f"Test accuracy: {test_accuracy:.4f}")
print(f"Test loss: {test_loss:.4f}")

# Step 12: Visualize training history
# Plot training & validation accuracy
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# Plot training & validation loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()

# Step 13: Make predictions and create confusion matrix
# Make predictions on test data
predictions = model.predict(test_generator, steps=test_steps)
predicted_classes = np.argmax(predictions, axis=1)

# Get true classes
true_classes = test_generator.classes[:len(predicted_classes)]

# Get class labels
class_labels = {v: k for k, v in test_generator.class_indices.items()}

# Create confusion matrix
cm = confusion_matrix(true_classes, predicted_classes)

# Get classification report
report = classification_report(true_classes, predicted_classes, target_names=list(class_labels.values()))
print("Classification Report:")
print(report)

# Step 14: Visualize confusion matrix
# For better visualization, we'll show just the top classes
def plot_confusion_matrix(cm, class_labels, num_classes_to_show=20):
    """
    Plots a confusion matrix for the top N most represented classes.
    """
    # Select the top N classes
    if len(class_labels) > num_classes_to_show:
        # Sum the confusion matrix across rows and columns to find the most represented classes
        class_representation = np.sum(cm, axis=0) + np.sum(cm, axis=1)
        top_classes_idx = np.argsort(class_representation)[-num_classes_to_show:]

        # Filter confusion matrix and class labels
        filtered_cm = cm[np.ix_(top_classes_idx, top_classes_idx)]
        filtered_labels = [class_labels[i] for i in top_classes_idx]
    else:
        filtered_cm = cm
        filtered_labels = list(class_labels.values())

    # Plot the confusion matrix
    plt.figure(figsize=(12, 10))
    sns.heatmap(filtered_cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=filtered_labels, yticklabels=filtered_labels)
    plt.title('Confusion Matrix (Top Classes)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Plot confusion matrix
plot_confusion_matrix(cm, class_labels)

# Step 15: Visualize some predictions
# Get a batch of test images
test_images, test_labels = next(test_generator)
num_samples = 10

# Make predictions
predictions = model.predict(test_images[:num_samples])
predicted_classes = np.argmax(predictions, axis=1)

# Get true class numbers
true_classes = np.argmax(test_labels[:num_samples], axis=1)

# Plot images with predictions
plt.figure(figsize=(15, 10))
for i in range(num_samples):
    plt.subplot(2, 5, i+1)
    plt.imshow(test_images[i])
    true_label = class_labels[true_classes[i]]
    pred_label = class_labels[predicted_classes[i]]
    title_color = 'green' if true_classes[i] == predicted_classes[i] else 'red'
    plt.title(f"True: {true_label}\nPred: {pred_label}", color=title_color)
    plt.axis('off')
plt.tight_layout()
plt.show()

# Step 16: Create and train a baseline model for comparison
def create_baseline_model(input_shape=(100, 100, 3), num_classes=131):
    """
    Creates a simple CNN baseline model without attention mechanisms.

    Args:
        input_shape: Shape of input images
        num_classes: Number of classes to classify

    Returns:
        The compiled model
    """
    model = models.Sequential([
        # First convolutional block
        Conv2D(32, (3, 3), padding='same', input_shape=input_shape),
        BatchNormalization(),
        Activation('relu'),
        Conv2D(32, (3, 3), padding='same'),
        BatchNormalization(),
        Activation('relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),

        # Second convolutional block
        Conv2D(64, (3, 3), padding='same'),
        BatchNormalization(),
        Activation('relu'),
        Conv2D(64, (3, 3), padding='same'),
        BatchNormalization(),
        Activation('relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),

        # Third convolutional block
        Conv2D(128, (3, 3), padding='same'),
        BatchNormalization(),
        Activation('relu'),
        Conv2D(128, (3, 3), padding='same'),
        BatchNormalization(),
        Activation('relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),

        # Flatten and fully connected layers
        layers.Flatten(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])

    return model

# Create baseline model
baseline_model = create_baseline_model(
    input_shape=(img_height, img_width, 3),
    num_classes=num_classes
)

# Compile baseline model
baseline_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train baseline model (fewer epochs to save time)
baseline_start_time = time.time()
baseline_history = baseline_model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,  # Fewer epochs for the baseline
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size
)
baseline_end_time = time.time()
baseline_training_time = baseline_end_time - baseline_start_time

# Evaluate baseline model
baseline_test_loss, baseline_accuracy = baseline_model.evaluate(test_generator, steps=test_steps)

# Step 17: Compare models
print("\nModel Comparison:")
print(f"Attention-Augmented EfficientNet - Test Accuracy: {test_accuracy:.4f}, Training Time: {training_time:.2f}s")
print(f"Baseline CNN - Test Accuracy: {baseline_accuracy:.4f}, Training Time: {baseline_training_time:.2f}s")
print(f"Improvement: {(test_accuracy - baseline_accuracy) * 100:.2f}%")

# Step 18: Save the final model
model.save(f"final_fruits_model_{timestamp}.h5")
print(f"Model saved as final_fruits_model_{timestamp}.h5")
```
